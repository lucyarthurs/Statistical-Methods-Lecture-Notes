# Probability Distributions

# 1. Probability distributions 

These can be either discrete (e.g bernounnli, binomial, poisson) or continuous (e.g. uniform, bernoulli, normal), and are defined by a mass/density function $ p(x) $ or $f(x)$.

## Probability distributions in R   

* ### Prefix "r" - randon number generation

* ### Prefix "p" - (cumulative) probability: F_X(x) = P(X <= x)

* ### Prefix "d" - density (Lebesgue density or counting density)

* ### Prefix "q" - quantile (generalized inverse of the cdf.)

## Commands are built by Prefix + Name of the distribution


## 1.1 Bernoulli distribution Be(p)

Flip a coin $(T = 0,  H = 1)$. The probability of `H` is 0.1

```r
options(repr.plot.width = 8, repr.plot.height = 8)

x <- 0:1
f <- dbinom(x, size = 1, prob = .3)
# plotting the distribution
plot(x, f, xlab = "x", ylab = "density", type = "h", lwd = 5, col = 'mediumpurple')
```

## 1.3 Binomial distribution binom(n,p)


```r
x <- 0:10 
y <- dbinom(x,10,.2) 
plot(x,y,type="h",xlab = "x",ylab="p(x)",main="pmf of a binom(n=10,p=0.2)") 
points(x,y,pch=19)
```

We have used the type="h" option to draw lines from the y-axis to each point in the plot. We used the points function with the pch=19 option to draw black dots at top of each line.

The cumulative distribution function is useful for 
 computing probabilities that a binomial random variable 
lies in a range. 
For example, suppose we needed to compute the probability 
that a binomial random variable X with success probability .2 
lies between 2 and 8. Actually, that is

Pr(2 < X < 8) = Pr(2 < X < 7) = Pr( X < 7) - Pr( X < 2)




```r
#To calculate with R we can just type

pbinom(7,10,.2)-pbinom(2,10,.2)
```

## 1. 3 Nomal distribution

The data values are members of a normally distributed population with mean $\mu$ and  variance $\sigma^2$.

The value of the distribution function is given by $P(X \leq x)$, the probability of the population to have values smaller than or equal to $x$.

```r
options(repr.plot.width = 8, repr.plot.height = 8)

x <- seq(-4,4,.1)
f <- dnorm(x, mean = 0, sd = 1)

# plotting
plot(x, f, xlab = "x", ylab = "density", lwd = 5,type = "l")
arrows(1, 0.05, 1, 0.1, lwd = 3, col = 'slateblue')
text(x = 0, y = 0.02, label = 'Note the area under the curve \n is the probability of
having an observation  between 0 and 2.')

abline(v = c(0,2), col = 'grey', lty = 5)
```

## 1.5 Overview

For practical computations R has built-in functions for the binomial, normal, Chi-squared distributions, among others. Where `d` stands for density, `p` for (cumulative) probability distribution, `q` for quantiles, and `r` for drawing random samples e.g.


| Distribution | parameters    | density                 | distributon             | random sampling          | quantiles                     |
|--------------|---------------|-------------------------|-------------------------|--------------------------|-------------------------------|
| Binomial     | n, p          | dbinom(x, n, p)         | pbinom(x, n, p)         | rbinom(10, n, p)         | qbinom($\alpha$, n, p)        |
| Normal       | $\mu, \sigma$ | dnorm(x, $\mu, \sigma$) | pnorm(x, $\mu, \sigma$) | rnorm(10, $\mu, \sigma$) | qnorm($\alpha$,$\mu, \sigma$) |
| Chi-squared  | m             | dchisq(x, m)            | pchisq(x, m)            | rchisq(10,m)             | qchisq($\alpha$, m)           |

## 1.6 uniform distribution

Although the probability density function is extremely simple for the uniform distribution, R provides a built-in function to compute it: dunif(x,a,b)


```r
#uniform distribution between 4 and 6
par(pch=16, mar=c(4,4,4,1)+0.1, pin=c(4,0.4))
a <- 4 # Lower bound
b <- 6 # Upper bound
h <- 1/(b-a) # Height of pdf
x <- c(a,a,b,b) # X-values of rectangle vertices
y <- c(0,h,h,0) # Y-values of rectangle vertices

plot(NULL, type="n", xlim=c(0,10), ylim=c(0,1), axes=FALSE, xlab= "Sample space",ylab="Density", main="Uniform Probability Density Function") # Empty plot
axis(1,at=seq(0,10)) # X-axis
axis(2,at=seq(0,1,0.5),las=1) # Y-axis
polygon(x,y,bor=NA,col="royalblue") # Polygon
text(a,h,bquote(italic(a)==.(a)),pos=3,xpd=TRUE) # Label a
text(b,h,bquote(italic(b)==.(b)),pos=3,xpd=TRUE) # Label b
```

```r
x=seq(0,10,0.1)
f=dunif(x,4,6)

plot(x, f, xlab = "x", ylab = "density", lwd = 5,type = "l", main="Uniform Density on [4,6]")
```

```r
par(pch=16, mar=c(4,4,4,1)+0.1, pin=c(4,0.4))
a <- 4 # Lower bound
b <- 6 # Upper bound
x <- c(a,b,b) # X-values of triangle vertices
y <- c(0,1,0) # Y-values of triangle vertices

plot(NULL, type="n", xlim=c(0,10), ylim=c(0,1), axes=FALSE, xlab= "Sample space",ylab="Density", main="Uniform CDF") # Empty plot
axis(1,at=seq(0,10)) # X-axis
axis(2,at=seq(0,1,0.5),las=1) # Y-axis
polygon(x,y,bor=NA,col="royalblue") # Polygon
text(a,0,bquote(italic(a)==.(a)),pos=3,xpd=TRUE) # Label a
text(b,1,bquote(italic(b)==.(b)),pos=3,xpd=TRUE) # Label b
```

## 2.1 Quantiles

###  (Theoretical) quantiles:
The *p-quantile* is the value with the property that there is a probability *p* of getting  a value less than or equal to it.

```r
q90<-qnorm(.90, mean = 0, sd = 1)
x<-seq(-4,4,.1)
f<-dnorm(x, mean = 0, sd = 1)
```

```r
# plotting
plot(x, f, xlab = "x", ylab = "density", type = "l", lwd = 4)
abline(v = q90, col = 'mediumpurple', lwd = 5)
arrows(2.5, 0.35, 1.5, 0.35, lwd = 3, col = 'midnightblue')
text(x = 2.8, y = 0.3, label ='90% of the prob \n is on the left of the 
purple vertical line');
```

### Empirical quantiles:

The *p-quantile* is the value with the property that p%  of the observations are less than or equal to it.

They can be easily obtained in R:

```r
x<-rnorm(100, mean = 0, sd = 1)
quantile(x)
```

```r
quantile(x,probs = c(.1,.2,.9))
```

## 2.4 QQ plot

Many statistical methods make some assumptions about the distribution of the data.

The quantile quantile (QQ) plot provides a mean to visually verify such assumptions. 

Also, the QQ-plot shows the theoretical quantile versus the empirical quantiles. If the distribution assumed (theoreticall) is indeed correct, the result will be a straight line. 

```r
set.seed(100)
x<-rnorm(100, mean = 0, sd = 1)
qqnorm(x)
qqline(x, col = 'mediumpurple', lwd = 3)
```

```r
require(graphics)

y <- rt(200, df = 5)
qqnorm(y); qqline(y, col = 2)
qqplot(y, rt(300, df = 5))

qqnorm(precip, ylab = "Precipitation [in/yr] for 70 US cities")

## "QQ-Chisquare" : --------------------------
y <- rchisq(500, df = 3)
## Q-Q plot for Chi^2 data against true theoretical distribution:
qqplot(qchisq(ppoints(500), df = 3), y,
       main = expression("Q-Q plot for" ~~ {chi^2}[nu == 3]))
qqline(y, distribution = function(p) qchisq(p, df = 3),
       prob = c(0.1, 0.6), col = 2)
mtext("qqline(*, dist = qchisq(., df=3), prob = c(0.1, 0.6))")
```

```r
set.seed(100)
x<-runif(100)
qqnorm(x)
qqline(x, col = 'mediumpurple', lwd = 3)
```

Note this is valid only for normal distributions!
